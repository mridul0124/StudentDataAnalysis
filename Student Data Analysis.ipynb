# Drop unnamed column

# Gender Distribution

```python
plt.figure(figsize=(4,5)) # figure size
ax = sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0]) # To display count
plt.title("Gender distribution.")
plt.show()
```

```python
# number of females are more than males.
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb)
```

# whenever we have single string colum and multiple integer column, 
# in that case we can use heatmap.

```python
sns.heatmap(gb)
plt.show()
```

```python
sns.heatmap(gb, annot=True)
plt.title("Parent's education qualification and student's scores.")
plt.show()
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1, annot=True)
plt.title("Parent Martial Status and Student Score")
plt.show()
```

```python
# Neglibible impact of parent's martial status on student's score.
```

## Distribution of ethnic group.

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = (df['EthnicGroup'] == "group A").count()
print(groupA)
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()


mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
label = ["group A","group B", "group C", "group D", "group E"]
print(mylist)
plt.pie(mylist, labels=label, autopct="%1.2f%%")
plt.title("Distribution of ethnic groups.")
plt.show()
```
